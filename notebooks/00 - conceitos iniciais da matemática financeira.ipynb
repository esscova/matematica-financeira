## 2. Representação e Metodologia: Diagrama de Fluxo de Caixa
A matemática financeira utiliza o Diagrama de Fluxo de Caixa (DFC) como um representação esquemática para organizar pagamentos e recebimentos ao longo de uma linha do tempo.



* Seta para baixo (Valores Negativos): Desembolso, saída de dinheiro ou investimento.
* Seta para cima (Valores Positivos): Ingresso, entrada de dinheiro ou resgate.
* Datas (x0,x1,xn): Indicam o momento exato da movimentação.

## 3. Operaçções de Avanço no Tempo: Juros e Capitalização
Quando um capital avança na linha do tempo, ele aumenta de valor devido aos juros. 

* Capital (C): Quantia inicial (Valor Presente).
* Montante (M): Quantia final (Valor Futuro).
* Juros (J): Remuneração do capital.

A relação fundamental é: 
$$
J=M−C 
$$

Ex.: Imagine que foi feita uma aplicação com Capital Inicial (Valor Presente) de R\$ 2.000,00 no resgate o Montante (Valor Futuro) foi de R\\$ 2.600,00, quanto foi a Remuneração(juros) da aplicação?

$$
\begin{align*}
    J &= 2600 - 2000 \\
    J &= 600
\end{align*}
$$

Independente do regime (juros simples ou compostos) o juros é a diferença entre o montante e o capital.

Vamos implementar isso em Python:
     

```python
def calcular_juros(montante:float, capital:float) -> float:
    """Calcula o Juros: J = M - C"""
    return montante - capital
```

```python
# capital
C = 2000

# montante
M = 2600

# juros
J = calcular_juros(M, C)
```

```python
# exibindo relatorio
print(f"Capital:  R$ {C:.2f}")
print(f"Montante: R$ {M:.2f}")
print(f"Juros:    R$ {J:.2f}")
```

## 4. Taxa de Juros (*i*)
A taxa de juros é a relação (divisão) entre os juros recbidos/pagos e o capital inicial.

Sua representação se faz:

$$
i = \frac{J}{C}
$$

Voltando ao exemplo anterior, vamos descobrir qual foi o valor da taxa unitária de juros.

$$
\begin{align*}
    i &= \frac{600}{2000} \\ \\
    i &= 0,3
\end{align*}
$$

Atenção aos formatos: 

     Taxa Unitária: Decimal (ex: 0,3). Indica quanto rende cada R$ 1,00.
     Taxa Percentual: Porcentagem (ex: 30%).
     Referencial Temporal: A taxa precisa de um prazo (a.m., a.a., etc.).

Vamos ao Python:

```python
def calcular_taxa(juros:float, capital:float) -> float:
    """Calcula a Taxa Unitária: i = J/C"""
    return juros/capital
```

```python
# calculando a taxa  unitaria
i_unitaria = calcular_taxa(J, C)

# calculando a taxa percentual
i_percentual = i_unitaria * 100 
```

```python
# relatorio
print(f"Taxa Unitária: {i_unitaria:.4f}") 
print(f"Taxa Percentual: {i_unitaria:.2%}") 
```

```python
# criando uma representação para o referencial temporal
referencial = "a.b." # biênio
print(f"Taxa Final: {i_unitaria:.2%} {referencial}")
```

## 5. Operações de Retrocesso no Tempo: Desconto
O desconto ocorre quando antecipamos um valor futuro. A relação matemática é a mesma dos juros, mas com nomes diferentes: 

$$
D=N−A 
$$  

     Valor Nominal (N): Valor de face (futuro).
     Valor Atual (A): Valor presente (antecipado).
     Desconto (D): Redução sofrida.

Para exemplificar, imagine que para antecipar uma duplicata no valor de R\\$10.000,00 o banco lhe oferece R$9.000,00, calculando o desconto vemos:

$$
\begin{align*}
    D &= 10000 - 9000 \\ \\
    D &= 1000
\end{align*}
$$

Seguindo, veremos sua aplicação no Python:

```python
def calcular_desconto(valor_nominal:float, valor_atual:float) -> float:
    """Calcula o Desconto: D = N - A"""
    return valor_nominal - valor_atual
```

```python
# valor nominal
N = 10000

# valor presente
A = 9000

# desconto
D = calcular_desconto(N, A)
```

```python
# relatorio
print(f"Valor Nominal: R$ {N:.2f}")
print(f"Valor Atual:   R$ {A:.2f}")
print(f"Desconto:      R$ {D:.2f}")
```
