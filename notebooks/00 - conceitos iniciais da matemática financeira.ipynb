# Conceitos Iniciais de Matemática Financeira
Os conceitos fundamentais da matemática financeira, tem como foco a dinâmica do valor do dinheiro ao longo do tempo.

A premissa central é que o dinheiro possui um valor variável dependendo da data em que é considerado, sendo remunerado por juros quando avança no tempo ou reduzido por descontos quando é antecipado.
