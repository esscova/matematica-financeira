## 2. Representação e Metodologia: Diagrama de Fluxo de Caixa
A matemática financeira utiliza o Diagrama de Fluxo de Caixa (DFC) como um representação esquemática para organizar pagamentos e recebimentos ao longo de uma linha do tempo.



* Seta para baixo (Valores Negativos): Desembolso, saída de dinheiro ou investimento.
* Seta para cima (Valores Positivos): Ingresso, entrada de dinheiro ou resgate.
* Datas (x0,x1,xn): Indicam o momento exato da movimentação.

## 3. Operaçções de Avanço no Tempo: Juros e Capitalização
Quando um capital avança na linha do tempo, ele aumenta de valor devido aos juros. 

* Capital (C): Quantia inicial (Valor Presente).
* Montante (M): Quantia final (Valor Futuro).
* Juros (J): Remuneração do capital.

A relação fundamental é: 
$$
J=M−C 
$$

Ex.: Imagine que foi feita uma aplicação com Capital Inicial (Valor Presente) de R\$ 2.000,00 no resgate o Montante (Valor Futuro) foi de R\\$ 2.600,00, quanto foi a Remuneração(juros) da aplicação?

$$
\begin{align*}
    J &= 2600 - 2000 \\
    J &= 600
\end{align*}
$$

Independente do regime (juros simples ou compostos) o juros é a diferença entre o montante e o capital.

Vamos implementar isso em Python:
     

```python
def calcular_juros(montante:float, capital:float) -> float:
    """Calcula o Juros: J = M - C"""
    return montante - capital
```

```python
# capital
C = 2000

# montante
M = 2600

# juros
J = calcular_juros(M, C)
```

```python
# exibindo relatorio
print(f"Capital:  R$ {C:.2f}")
print(f"Montante: R$ {M:.2f}")
print(f"Juros:    R$ {J:.2f}")
```

## 4. Taxa de Juros (*i*)
A taxa de juros é a relação (divisão) entre os juros recbidos/pagos e o capital inicial.

Sua representação se faz:

$$
i = \frac{J}{C}
$$

Voltando ao exemplo anterior, vamos descobrir qual foi o valor da taxa unitária de juros.

$$
\begin{align*}
    i &= \frac{600}{2000} \\ \\
    i &= 0,3
\end{align*}
$$

Atenção aos formatos: 

     Taxa Unitária: Decimal (ex: 0,3). Indica quanto rende cada R$ 1,00.
     Taxa Percentual: Porcentagem (ex: 30%).
     Referencial Temporal: A taxa precisa de um prazo (a.m., a.a., etc.).

Vamos ao Python:

```python
def calcular_taxa(juros:float, capital:float) -> float:
    """Calcula a Taxa Unitária: i = J/C"""
    return juros/capital
```

```python
# calculando a taxa  unitaria
i_unitaria = calcular_taxa(J, C)

# calculando a taxa percentual
i_percentual = i_unitaria * 100 
```

```python
# relatorio
print(f"Taxa Unitária: {i_unitaria:.4f}") 
print(f"Taxa Percentual: {i_unitaria:.2%}") 
```

```python
# criando uma representação para o referencial temporal
referencial = "a.b." # biênio
print(f"Taxa Final: {i_unitaria:.2%} {referencial}")
```

## 5. Operações de Retrocesso no Tempo: Desconto
O desconto ocorre quando antecipamos um valor futuro. A relação matemática é a mesma dos juros, mas com nomes diferentes: 

$$
D=N−A 
$$  

     Valor Nominal (N): Valor de face (futuro).
     Valor Atual (A): Valor presente (antecipado).
     Desconto (D): Redução sofrida.

Para exemplificar, imagine que para antecipar uma duplicata no valor de R\\$10.000,00 o banco lhe oferece R$9.000,00, calculando o desconto vemos:

$$
\begin{align*}
    D &= 10000 - 9000 \\ \\
    D &= 1000
\end{align*}
$$

Seguindo, veremos sua aplicação no Python:

```python
def calcular_desconto(valor_nominal:float, valor_atual:float) -> float:
    """Calcula o Desconto: D = N - A"""
    return valor_nominal - valor_atual
```

```python
# valor nominal
N = 10000

# valor presente
A = 9000

# desconto
D = calcular_desconto(N, A)
```

```python
# relatorio
print(f"Valor Nominal: R$ {N:.2f}")
print(f"Valor Atual:   R$ {A:.2f}")
print(f"Desconto:      R$ {D:.2f}")
```

## 6. Simetria entre Juros e Desconto (Orientação a Objetos)
Juros e Desconto são o mesmo fenômeno financeiro vistos de ângulos opostos. 

| Aspecto              | Operação de Juros           | Operação de Desconto          |
| -------------------- | --------------------------- | ----------------------------- |
| **Sentido no Tempo** | Avançar (Progredir)         | Retroceder (Antecipar)        |
| **Valor Inicial**    | Capital Inicial / Principal | Valor Atual / Valor Presente  |
| **Valor Final**      | Montante / Valor Futuro     | Valor Nominal / Valor de Face |
| **Remuneração**      | Juros (Acréscimo)           | Desconto (Dedução)            |
| **Efeito no Valor**  | O capital aumenta           | O capital diminui             |

A orientação a objetos (OOP) em Python é perfeita para representar essa simetria, criando uma estrutura única que se adapta ao sentido do tempo.

```python
class OperacaoFinanceira:
    def __init__(self, valor_inicial:float, valor_final:float, tipo:str='capitalização'):
        self.valor_inicial = valor_inicial
        self.valor_final = valor_final
        self.tipo = tipo

    def calcular_remuneracao(self) -> float:
        """Retorna Juros ou Desconto"""
        return self.valor_final - self.valor_inicial

    def calcular_taxa(self) -> float:
        """Calcula a taxa sobre o valor inicial"""
        return self.calcular_remuneracao() / self.valor_inicial

    def gerar_relatorio(self) -> None:
        remuneracao = self.calcular_remuneracao()
        taxa = self.calcular_taxa()

        if self.tipo == 'capitalização':
            print("📊 OPERAÇÃO DE JUROS (Avanço no Tempo)")
            print(f"Capital (C)......: R$ {self.valor_inicial:.2f}")
            print(f"Montante (M).....: R$ {self.valor_final:.2f}")
            print(f"Juros (J)........: R$ {remuneracao:.2f}")
        else:
            print("📉 OPERAÇÃO DE DESCONTO (Retrocesso no Tempo)")
            print(f"Valor Atual (A)..: R$ {self.valor_inicial:.2f}")
            print(f"Valor Nominal (N): R$ {self.valor_final:.2f}")
            print(f"Desconto (D).....: R$ {remuneracao:.2f}")
            
        print(f"Taxa.............: {taxa:.2%}")
```

Tendo a classe que nos auxiliará construída voltamos aos exemplos anteriores:
1. A aplicação com Capital Inicial (Valor Presente) de R\\$ 2.000,00 com resgate de Montante (Valor Futuro) no valor de R\\$ 2.600,00 e,
2. A antecipação de duplicata no valor de R\\$10.000,00 onde o banco oferece R$9.000,00 finais.

```python
# instanciando os casos
print("="*40)
op_1 = OperacaoFinanceira(2000, 2600, 'capitalização')
op_1.gerar_relatorio()

print("="*40)
op_2 = OperacaoFinanceira(9000, 10000, "desconto")
op_2.gerar_relatorio()
```

## 7. Visualização: Desenhando o Diagrama de Fluxo de Caixa 
Vamos usar a biblioteca `matplotlib` para transformar dados em DFCs visuais.

```python
def plotar_dfc(fluxos_caixa:list, titulo:str='Diagrama de Fluxo de Caixa'):
    """
    Desenha um DFC.
    Valor negativo = saída (seta para baixo)
    Valor positivo = entrada (seta para cima)
    Args:
        fluxos_caixa: lista de tuplas [(período, valor), ...]
        titulo:str = "Diagrama de Fluxo de Caixa" por padrão
    """

    periodos = [fc[0] for fc in fluxos_caixa]
    valores = [fc[1] for fc in fluxos_caixa]

    fig, ax = plt.subplots(figsize=(10,4))

    # timeline
    ax.hlines(
        0, 
        xmin=min(periodos) - .5,
        xmax=max(periodos) + .5,
        color='black',
        linewidth=1
        )

    # setas
    for p, v in fluxos_caixa:
        cor = 'green' if v > 0 else 'red'
        direcao = 1 if v > 0 else -1


        # haste + ponta
        ax.vlines(p, ymin=0, ymax=v, colors=cor, linewidth=2)
        ax.scatter(p, v, color=cor, zorder=5, marker=(3, 0, 180 if v < 0 else 0), s=100)

        # texto
        offset = 100 * direcao
        ax.text(p, v + offset, f"R$ {abs(v):.0f}", ha='center', color=cor, fontweight='bold')
        ax.text(p, -300, f"x_{p}", ha='center')

    ax.set_title(titulo, fontsize=14)
    ax.get_yaxis().set_visible(False) # esconder eixo Y
    plt.show()        
```

```python
# desenhando o Caso 1: Investimento no tempo 0, Resgate no tempo 2
dfc_caso_a = [(0, -2000), (2, 2600)]
plotar_dfc(dfc_caso_a, "Caso 1: Capitalização (Juros)")
```

```python
# Desenhando o Caso 2: Recebimento no tempo 0, Pagamento no tempo 1
dfc_caso_b = [(0, 9000), (1, -10000)]
plotar_dfc(dfc_caso_b, "Caso 1: Operação de Desconto")
```
